# 1. Data description

Initially, two datasets describing temporal behaviour of bitcoin are used. The main approach is to build a temporal and a regression model, and compare their predictive power on the cryptocurrency's price.

## First dataset - only price variations.

This dataset is available at https://www.kaggle.com/datasets/varpit94/bitcoin-data-updated-till-26jun2021 It contains the following columns:
- Date
- Open - Price from the first transaction of a trading day.
- High - Maximum price in a trading day.
- Low - Minimum price in a trading day.
- Close - Price from the last transaction of a trading day.
- Adj Close - Closing price adjusted to reflect the value after accounting for any corporate actions.
- Volume - Number of units traded in a day.



```python
df_price = pd.read_csv('../data/BTC-USD.csv', parse_dates=True)
```

```python
df_price.head()
```

```python
df_price.info()
```

```python
df_price.describe()
```

```python
df_price = df_price.set_index('Date')
df_price.index = pd.to_datetime(df_price.index)
```

```python
df_price
```

```python
print(pd.date_range(start="2014-09-17", end="2022-03-25").difference(df_price.index))
```

Conclusion: no missing days in the dataset between 17.09.2014 and 25.03.2022.

```python
df_price.drop(columns='Volume').plot()
plt.title('BTC price over time')
plt.ylabel('Price (USD)')
```

```python
df_price.hist(bins=100)
```

```python
df_price['Amplitude'] = df_price.High - df_price.Low
```

```python
df_price['Amplitude'].rolling(7).mean
```

```python
plt.plot(df_price.index, df_price['Amplitude'], label='Data')
plt.plot(df_price.index, df_price['Amplitude'].rolling(7).mean(), label='7-day running mean')
plt.xlabel('Date')
plt.ylabel('USD')
plt.title('Daily price amplitude of BTC')
plt.legend()
```

```python
df_price['Change'] = df_price.Close - df_price.Open
```

```python
plt.plot(df_price.index, df_price['Change'], label='Data')
plt.plot(df_price.index, df_price['Change'].rolling(7).mean(), label='7-day running mean')
plt.xlabel('Date')
plt.ylabel('USD')
plt.title('Daily price change of BTC')
plt.legend()
```

```python
plt.plot(df_price.index, df_price['Volume'],  label='Data')
plt.plot(df_price.index, df_price['Volume'].rolling(7).mean(), label='7-day running mean')
plt.xlabel('Date')
plt.yscale('log')
plt.title('Daily number of transactions')
plt.legend()
```

```python
df_price.loc[:, ['Amplitude', 'Change']].hist(bins=100)
```

## Second dataset - price + additional market information

This dataset is available at https://datahub.io/cryptocurrency/bitcoin#data-cli It contains the following columns:

- txVolume(USD) - on-chain transaction volume.
- txCount - refers to the number of transactions happening on the public blockchain a day.
- marketcap(USD) - This is the unit price multiplied by the number of units in circulation.
- price - opening price per day.
- exchangevolume(USD) - the dollar value of the volume at exchanges.
- generatedCoins - the number of new coins that have been brought into existence on that day.
- fees - fees for using the blockchain based on the native currency (BTC)
- averaDifficulty - average measure of how difficult it is to find a hash below a given target.

As around 2000 first rows contain lots of null values we decided to drop them completely and focus on the later period.

```python
df_market = pd.read_csv('../data/bitcoin_csv.csv', parse_dates=True).dropna()
```
