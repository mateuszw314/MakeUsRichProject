In this notebook we would like to use a Bayesian neural network as a regression tool on the *Datahub* [Bitcoin dataset](https://datahub.io/cryptocurrency/bitcoin). We will be following a [Keras tutorial](https://keras.io/examples/keras_recipes/bayesian_neural_networks/). We were quite happy with the results from notebook *Bitcoin_regression.ipynb*, however  we are concerned about the size of our dataset, as it, after removing empty rows, has data from around 2000 consecutive days only and also we are not sure about how much noise is in our dataset. Bayesian neural networks tend to deal with such problems better than other learning algorithms, that's why we will try to use one for the Bitcoin price regression problem.

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import tensorflow_datasets as tfds
import tensorflow_probability as tfp
import pandas as pd
import datetime
from datapackage import Package
from matplotlib import pyplot as plt
import itertools
import seaborn as sns
from sklearn.feature_selection import SelectKBest, f_regression

# matplotlib options
palette = itertools.cycle(sns.color_palette())
plt.style.use('ggplot')
%matplotlib inline
plt.rcParams['figure.figsize'] = (16, 10)
```

Then, we define a helper function to load our dataset properly. The constant *FEATURE_NAMES* contains the names of our features

```python
FEATURE_NAMES = [
'txVolume(USD)', 'adjustedTxVolume(USD)', 'txCount', 'marketcap(USD)', 'exchangeVolume(USD)', 'generatedCoins','fees', 'activeAddresses', 'averageDifficulty', 'paymentCount', 'medianTxValue(USD)', 'medianFee','blockSize', 'blockCount'
]

def array_to_dict(array):
  newDict = {}
  for count, value in enumerate(FEATURE_NAMES):
    newDict[value]=array[count]
  return newDict
```

Then we load our dataset, while doing some data preprocessing. For example, we remove all the data rows until until 2013-04-28, getting rid of first 1571 rows, as they were mostly empty. We naturally standardize the dataset.

```python
  # Loading bitcoin dataset
def load_dataset():
    
    df=[]

    # print processed tabular data (if exists any)
    for resource in package.resources:
        if resource.descriptor['datahub']['type'] == 'derived/csv':
            df=resource.read()

    df=np.asarray(df)
    df=df[1571:]
    df=np.delete(df,obj=0,axis=1)
    prices = df[:,4]
    prices = prices.astype(float)
    df=np.delete(df,obj=4,axis=1)
    df=df.astype(float)

    # standardizing 
    df_mean = df.mean()
    df_std =df.std()
    df = (df - df_mean) / df_std

    prices_mean = prices.mean()
    prices_std = prices.std()
    prices = (prices - prices_mean) / prices_std


    dataset = tf.data.Dataset.from_tensor_slices((df, prices))
    dataset=dataset.map(lambda x,y: (array_to_dict(x),y))
    return dataset, prices_mean, prices_std

package = Package('https://datahub.io/cryptocurrency/bitcoin/datapackage.json')
dataset,prices_mean,prices_std=load_dataset()
```

Let's define a function that splits our dataset into the train and test subsets.

```python
def get_train_and_test_splits(train_size, dataset, batch_size=1):
    train_dataset = (
        dataset.take(train_size).shuffle(buffer_size=train_size).batch(batch_size)
    )
    test_dataset = dataset.skip(train_size).batch(batch_size)

    return train_dataset, test_dataset
```

Now we define a function that creates one Input layer of our neural network per feature

```python
def create_model_inputs():
    inputs = {}
    for feature_name in FEATURE_NAMES:
        inputs[feature_name] = layers.Input(
            name=feature_name, shape=(1,), dtype=tf.float32
        )

    return inputs
```

Now we specify prior and posterior weight distributions, but we leave the prior distribution not trainable with fixed parameters. The prior distribution is *Normal(0,1)* while the posterior is a *MultivariateGaussian* that will be learning its means, variances and covariances. We chose those distributions arbitrarily, being inspired by the *Keras* tutorial.

```python
# Define the prior weight distribution as Normal of mean=0 and stddev=1.
# Note that, in this example, the we prior distribution is not trainable,
# as we fix its parameters.
def prior(kernel_size, bias_size, dtype=None):
    n = kernel_size + bias_size
    prior_model = keras.Sequential(
        [
            tfp.layers.DistributionLambda(
                lambda t: tfp.distributions.MultivariateNormalDiag(
                    loc=tf.zeros(n), scale_diag=tf.ones(n)
                )
            )
        ]
    )
    return prior_model


# Define variational posterior weight distribution as multivariate Gaussian.
# Note that the learnable parameters for this distribution are the means,
# variances, and covariances.
def posterior(kernel_size, bias_size, dtype=None):
    n = kernel_size + bias_size
    posterior_model = keras.Sequential(
        [
            tfp.layers.VariableLayer(
                tfp.layers.MultivariateNormalTriL.params_size(n), dtype=dtype
            ),
            tfp.layers.MultivariateNormalTriL(n),
        ]
    )
    return posterior_model
```

Now we split our dataset into train and test subsets. By trial and error, we decided on the batch size to be 80.

```python
#split into train and test

dataset_size = len(dataset)
batch_size = 500
train_size = int(dataset_size * 0.9)
train_dataset, test_dataset = get_train_and_test_splits(train_size, dataset, batch_size)
mse_loss = keras.losses.MeanSquaredError()
hidden_units = [8, 8]
learning_rate = 0.001
```

Now let us define our bayesian neural network model. It has 14 input layers, that are then concatenated on the concatenation layer. Then it has 2 hidden layers with 8 units each. And then we have an output layer with one output unit. 

```python
def create_bnn_model(train_size):
    inputs = create_model_inputs()
    features = keras.layers.concatenate(list(inputs.values()))
    features = layers.BatchNormalization()(features)

    # Create hidden layers with weight uncertainty using the DenseVariational layer.
    for units in hidden_units:
        features = tfp.layers.DenseVariational(
            units=units,
            make_prior_fn=prior,
            make_posterior_fn=posterior,
            kl_weight=1 / train_size,
            activation="sigmoid"
        )(features)

    # The output is deterministic: a single point estimate.
    outputs = layers.Dense(units=1)(features)
    model = keras.Model(inputs=inputs, outputs=outputs)
    return model
```

We define a training and evaluating function with mean squared error loss function and also a root mean square error function to evaluate the model. 

```python
def train_and_evaluate(model, loss, train_dataset, test_dataset, num_epochs):

    model.compile(
        optimizer=keras.optimizers.RMSprop(learning_rate=learning_rate),
        loss=loss,
        metrics=[keras.metrics.RootMeanSquaredError()],
    )

    print("Start training the model...")
    model.fit(train_dataset, epochs=num_epochs, validation_data=test_dataset)
    print("Model training finished.")
    _, rmse = model.evaluate(train_dataset, verbose=0)
    print(f"Train RMSE: {round(rmse, 3)}")

    print("Evaluating model performance...")
    _, rmse = model.evaluate(test_dataset, verbose=0)
    print(f"Test RMSE: {round(rmse, 3)}")
```

By trial and error, we found the 1500 epochs to give the best results, along with the batch size of 500 and a learning rate of 0.001. The batch size might seem quite large, but regarding our dataset, with many irregular fluctuations one by one, we were hoping it would help the model grasp the fluctuation patters better. 

Now let us create the model, train and evaluate it. 

```python
# Create, train, evaluate
num_epochs = 1500
small_train_dataset = train_dataset.take(train_size)
bnn_model = create_bnn_model(train_size)
train_and_evaluate(bnn_model, mse_loss, small_train_dataset, test_dataset, num_epochs)
```

The test *RMSE* seems to be promisingly low, however the data still has to be de-standardized for us to be able to compare it with the results of *Bitcoin_regression.ipynb* 

Let us choose some samples and test the model more

Whenever we plot our graphs, predicted price values are in blue, while the actual price values are in red. 

```python
def plot_a_range(plot_range, preds, y_true, error_min,error_max):
  plt.errorbar(range(plot_range),preds[:plot_range],[error_min[:plot_range],error_max[:plot_range]],fmt='bo')
  plt.plot(y_true[:plot_range],'r+')

def compute_predictions(model, targets, examples,prices_std, prices_mean,iterations=100,plot_range=40):
  predicted = []
  for _ in range(iterations):
      predicted.append(model(examples).numpy())
  predicted = np.concatenate(predicted, axis=1)

  #convert back to original scale
  predicted = predicted * prices_std + prices_mean
  targets = targets * prices_std + prices_mean

  prediction_mean = np.mean(predicted, axis=1).tolist()
  prediction_min = np.min(predicted, axis=1).tolist()
  prediction_max = np.max(predicted, axis=1).tolist()
  prediction_range = (np.max(predicted, axis=1) - np.min(predicted, axis=1)).tolist()

  # plotting the means of predicted values and the actual values
  plot_a_range(plot_range,prediction_mean,targets,prediction_min,prediction_max)

sample = 100
examples, targets = list(test_dataset.unbatch().shuffle(batch_size * 10).batch(sample))[
    0
]

compute_predictions(bnn_model,targets,examples,prices_std, prices_mean)
```

It is not that trivial to compare the results of this model to our regression models from *Bitcoin_regression.ipynb* naturally, as this model produces different outputs for different inputs each time it is run. That's why we run the predictions multiple times, 100 by default, to plot the graph above. Even though the predicted price range is relatively big, and single predictions can be quite off (which suggests that model's mode weights are still quite uncertain), the means are relatively close to the real values, which proves that the network has learned to some extent and is doing relatively alright. 

What could have been further done here, is to try feature selection like in the previous notebook, play more with the parameters, or improve the model in some other way. We could also try with some different prior and posterior distributions, especially since we assumed the prior distribution to be *Normal(0,1)*.
